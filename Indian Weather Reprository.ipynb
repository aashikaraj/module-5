## Which city has the highest difference between temperature and feels like temperature in Celsius and what is the time there?

```python
data['temp_diff'] = data['temperature_celsius']-data['feels_like_celsius']
```

```python
high_diff_row = data.loc[data['temp_diff'].idxmax()]

location_name = high_diff_row['location_name']
region = high_diff_row['region']
timezone = high_diff_row['timezone']
time = high_diff_row['last_updated']
```

```python
print(f"The city with highest difference between temperature and feels like temperature in Celsius is {location_name},{region}({timezone})with a difference of {high_diff_row['temp_diff']}°C. The  time there was {time}.")
```

## What is the average 'temperature_celsius' across all locations?

```python
average = data['temperature_celsius'].mean()
print(f'The average temperature in celsius across all location is {average:.2f} degree.')
```

## Is there a correlation between 'temperature_celsius' and 'humidity'?

```python
corr = data['temperature_celsius'].corr(data['humidity'])
print(f'The correlation between temperature (Celsius) and humidity is {corr:.2f}')
```

```python
import seaborn as sns

# Select relevent columns for correlation analysis
correlation_cols = ['temperature_celsius', 'precip_mm', 'wind_kph', 'humidity', 'air_quality_PM2.5']

# Calculate the correlation matrix
correlation_matrix = data[correlation_cols].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
x = data['temperature_celsius']
y = data['humidity']

plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.5)
plt.title("Scatter Plot of Temperature (Celsius) vs. Humidity")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Humidity")
plt.grid(True)
plt.show()
```

### Insight : The correlation coefficient of 0.05 suggest a very weak or negligiblecorrelation between temperature and humidity

### Sun and Moon Data Questions:

#### What time does the sun usually set in the 'sunset' column for a specific location?

```python
location_name = 'Bikaner' 
location_data = data[data['location_name'] == location_name]
if not location_data.empty:
    location_data['sunset'] = pd.to_datetime(location_data['sunset'])
    average_sunset = location_data['sunset'].mean()
    
    print(f"The sun usually sets at {average_sunset.strftime('%H:%M:%S')} for {location_name}.")
else:
    print(f"No  Data Found for {location_name}.")
```

### The sun usually sets at 18:27:27 for Bikaner.

#### Is there any relationship between 'moon_phase' and 'temperature_celsius'?

```python
moon_phase = data['moon_phase']
celsius = data['temperature_celsius']

plt.figure(figsize=(8,6))
plt.scatter(moon_phase,celsius,alpha=0.5)
plt.title("Scatter plot of moon phase vs Temperature(celsius)")
plt.xlabel("Moon phase")
plt.ylabel("Temperature in celsius")
plt.xticks(rotation=45) 
plt.grid(True)
plt.show()
```

#### insights


#### Scatter Plot of Moon Phase vs Temperature

The scatter plot shows a weak correlation between moon phase and temperature, with temperatures being more variable than moon phases. This is likely due to the fact that temperature is affected by a number of factors, including sunlight, cloud cover, and wind speed, while moon phase is a much more regular phenomenon .

temperature is higher during the full moon and lower during the new moon
the temperature differnce between the full moon and new moon is also greater during the winter months
the scatter plot indicates that there is small but measurable correlation between moon phase and temperature
the data points are more scattered at higher temperature
